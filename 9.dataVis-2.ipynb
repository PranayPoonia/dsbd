```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df.isnull().sum()
```

```python
newdf=df[df['age']>15]
newdf=newdf[newdf['age']<40]
```

```python
sns.boxplot(x='sex', y="age", data=newdf, hue="survived")
```
